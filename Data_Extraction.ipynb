```python
import os
import json
import pandas as pd

# ====================== BASE PATH ======================
base_path = r"D:\Guvi Projects\PhonePe Transaction Insights\pulse\data\\"   # fixed

# ======================================================
# 1. AGGREGATED TRANSACTION
# ======================================================
print("Extracting 1/9 : agg_transaction")

path1 = base_path + "aggregated/transaction/country/india/state/"   # fixed slash consistency
agg_trans_list = os.listdir(path1)

columns1 = {'State':[], 'Year':[], 'Quarter':[], 'Transaction_type':[], 
            'Transaction_count':[], 'Transaction_amount':[]}

for state in agg_trans_list:
    cur_state = path1 + state + "/"
    if not os.path.isdir(cur_state):
        continue
    agg_year_list = os.listdir(cur_state)
    
    for year in agg_year_list:
        cur_year = cur_state + year + "/"
        if not os.path.isdir(cur_year):
            continue
        agg_file_list = os.listdir(cur_year)
        
        for file in agg_file_list:
            cur_file = cur_year + file
            if not os.path.isfile(cur_file):
                continue
                
            with open(cur_file, 'r', encoding='utf-8') as data:   # fixed file handling
                A = json.load(data)
            
            if A.get('data') and A['data'].get('transactionData'):   # fixed crash issue
                for i in A['data']['transactionData']:
                    name = i['name']
                    count = i['paymentInstruments'][0]['count']
                    amount = i['paymentInstruments'][0]['amount']
                    
                    columns1['State'].append(state)
                    columns1['Year'].append(year)
                    columns1['Quarter'].append(int(file.replace('.json', '')))   # fixed strip bug
                    columns1['Transaction_type'].append(name)
                    columns1['Transaction_count'].append(count)
                    columns1['Transaction_amount'].append(amount)

df_agg_trans = pd.DataFrame(columns1)
print("df_agg_trans shape :", df_agg_trans.shape)
```

```python
# ======================================================
# 4. MAP TRANSACTION (Latest Format - hoverDataList)
# ======================================================
print("\nExtracting 4/9 : map_transaction")

path4 = base_path + "map/transaction/hover/country/india/state/"

columns4 = {'State':[], 'Year':[], 'Quarter':[], 'District':[], 
            'Transaction_count':[], 'Transaction_amount':[]}

files_processed = 0

for state in os.listdir(path4):
    cur_state = path4 + state + "/"
    if not os.path.isdir(cur_state):
        continue

    for year in os.listdir(cur_state):
        cur_year = cur_state + year + "/"
        if not os.path.isdir(cur_year):
            continue

        for file in os.listdir(cur_year):
            if not file.endswith('.json'):
                continue
                
            cur_file = cur_year + file
            files_processed += 1
            
            with open(cur_file, 'r', encoding='utf-8') as data:
                A = json.load(data)
            
            try:
                if A.get('data') and A['data'].get('hoverDataList'):
                    hover_list = A['data']['hoverDataList']
                    
                    quarter = int(file.replace('.json', '').strip())
                    
                    for item in hover_list:
                        name = item.get('name')
                        metric = item.get('metric', [{}])[0]
                        
                        columns4['State'].append(state)
                        columns4['Year'].append(year)
                        columns4['Quarter'].append(quarter)
                        columns4['District'].append(name)
                        columns4['Transaction_count'].append(metric.get('count'))
                        columns4['Transaction_amount'].append(metric.get('amount'))
                        
            except:
                continue

print("Total files processed:", files_processed)
df_map_trans = pd.DataFrame(columns4)
print("df_map_trans shape :", df_map_trans.shape)
print("Sample data:\n", df_map_trans.head(5) if not df_map_trans.empty else "Still empty")
```

```python
# ======================================================
# 5. MAP USER (Fixed - Using hoverData dict)
# ======================================================
print("\nExtracting 5/9 : map_user")

path5 = base_path + "map/user/hover/country/india/state/"

columns5 = {'State':[], 'Year':[], 'Quarter':[], 'District':[], 
            'Registered_users':[], 'App_opens':[]}

files_processed = 0

for state in os.listdir(path5):
    cur_state = path5 + state + "/"
    if not os.path.isdir(cur_state):
        continue

    for year in os.listdir(cur_state):
        cur_year = cur_state + year + "/"
        if not os.path.isdir(cur_year):
            continue

        for file in os.listdir(cur_year):
            if not file.endswith('.json'):
                continue
                
            cur_file = cur_year + file
            files_processed += 1
            
            with open(cur_file, 'r', encoding='utf-8') as data:
                A = json.load(data)
            
            try:
                if A.get('data') and A['data'].get('hoverData'):
                    hover_dict = A['data']['hoverData']
                    
                    quarter = int(file.replace('.json', '').strip())
                    
                    for district, value in hover_dict.items():
                        columns5['State'].append(state)
                        columns5['Year'].append(year)
                        columns5['Quarter'].append(quarter)
                        columns5['District'].append(district)
                        columns5['Registered_users'].append(value.get('registeredUsers', 0))
                        columns5['App_opens'].append(value.get('appOpens', 0))
            except:
                continue

print("Total files processed:", files_processed)
df_map_user = pd.DataFrame(columns5)
print("df_map_user shape :", df_map_user.shape)
print("Sample data:\n", df_map_user.head(5) if not df_map_user.empty else "Still empty")
```

```python
# ======================================================
# 7. TOP TRANSACTION
# ======================================================
print("\nExtracting 7/9 : top_transaction")

path7 = base_path + "top/transaction/country/india/state/"

columns7 = {'State':[], 'Year':[], 'Quarter':[], 'Entity_type':[], 'Entity_name':[], 
            'Transaction_count':[], 'Transaction_amount':[]}

for state in os.listdir(path7):
    cur_state = path7 + state + "/"
    if not os.path.isdir(cur_state):
        continue

    for year in os.listdir(cur_state):
        cur_year = cur_state + year + "/"
        if not os.path.isdir(cur_year):
            continue

        for file in os.listdir(cur_year):
            if not file.endswith('.json'):
                continue
                
            cur_file = cur_year + file
            with open(cur_file, 'r', encoding='utf-8') as data:
                A = json.load(data)
            
            try:
                quarter = int(file.replace('.json', '').strip())
                data_top = A.get('data', {})
                
                # Districts
                for district in data_top.get('districts', []):
                    columns7['State'].append(state)
                    columns7['Year'].append(year)
                    columns7['Quarter'].append(quarter)
                    columns7['Entity_type'].append('District')
                    columns7['Entity_name'].append(district.get('entityName'))
                    metric = district.get('metric', {})
                    columns7['Transaction_count'].append(metric.get('count'))
                    columns7['Transaction_amount'].append(metric.get('amount'))
                
                # Pincodes
                for pin in data_top.get('pincodes', []):
                    columns7['State'].append(state)
                    columns7['Year'].append(year)
                    columns7['Quarter'].append(quarter)
                    columns7['Entity_type'].append('Pincode')
                    columns7['Entity_name'].append(pin.get('entityName'))
                    metric = pin.get('metric', {})
                    columns7['Transaction_count'].append(metric.get('count'))
                    columns7['Transaction_amount'].append(metric.get('amount'))
            except:
                continue

df_top_trans = pd.DataFrame(columns7)
print("df_top_trans shape :", df_top_trans.shape)
```
